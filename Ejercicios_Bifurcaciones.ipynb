# **Ejercicios Bifurcaciones**

Módulos de librerias

```
import matplotlib.style
matplotlib.style.use('classic')
import numpy as np
import time
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
```

## Ejercicio 3.4.2

\begin{equation}
 \frac{dx}{dt}=rx-sinh(x)
 \end{equation} 

```
x=np.arange(-10, 10, 0.01)

dxdt= 2*x-np.sinh(x)
fig = plt.figure()
plt.plot( x, dxdt, linewidth = 2)
plt.title(r'Gráfico $\frac{dx}{dt}$ vs $x$', fontsize=20)
px1 = 5
py1 = 2*px1-np.sinh(px1)
px2 = -5
py2 = 2*px2-np.sinh(px2)
punto = plt.plot([px1], [py1], 'bo')
nota = plt.annotate(r'$5$',xy=(px1, py1),xycoords='data',xytext=(5,1000),fontsize=14)
punto2 = plt.plot([px2], [py2], 'bo')
nota2 = plt.annotate(r'$-5$',xy=(px2, py2),xycoords='data',xytext=(-5,1000),fontsize=14)
plt.xlim([-11, 11])
#plt.ylim([2.5, 5.5])
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$\frac{dx}{dt}$', fontsize=20)
plt.grid()
plt.show()
```

```
def sistema_2(t, x,a,b):
    x_dot = a*x-np.sinh(x)
    return x_dot
```

```
b = 4,3
x0 = np.linspace(-10,10,10)
tspan = [0, 100]
r = np.linspace(-5,5,100)
tic = time.time()
for j in range(len(r)):
    sol = solve_ivp(sistema_2, tspan, x0,args=(r[j],b))
    t = sol.t
    x = sol.y
    for i in range(x0.size):
        plt.scatter(r[j],x[i,-1] , color='green', marker='.')

x0 = np.linspace(-.1,.1,5)
tspan = [0, -100]

r = np.linspace(1.1,5,50)
tic = time.time()
for j in range(r.size):
    sol = solve_ivp(sistema_2, tspan, x0,args=(r[j],b))
    t = sol.t
    x = sol.y
    for i in range(x0.size):
        plt.scatter(r[j],x[i,-1] , color='red', marker='.')
        plt.title('Diagrama de bifurcación')
        plt.xlabel('r')
        plt.ylabel('x')

toc = time.time()
plt.show()
```

## Ejercicio 3.4.3

\begin{equation}
 \frac{dx}{dt}=rx-4x^{3}
 \end{equation} 

```
x=np.arange(-10, 10, 0.01)

dxdt= (2*x)-(4*x**3)
fig = plt.figure()
plt.plot( x, dxdt, linewidth = 2)
plt.title(r'Gráfico $\frac{dx}{dt}$ vs $x$', fontsize=20)
px1 = 0
py1 = 2*px1-np.sinh(px1)
punto = plt.plot([px1], [py1], 'bo')
nota = plt.annotate(r'$0$',xy=(px1, py1),xycoords='data',xytext=(0,100),fontsize=14)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$\frac{dx}{dt}$', fontsize=20)
plt.grid()
plt.show()
```

```
def sistema_1(t, x,a,b):
    x_dot = (a*x)-(4*x**3)
    return x_dot
```

```
b = 4,3
x0 = np.linspace(-10,10,10)
tspan = [0, 100]
r = np.linspace(-2,2,100)
tic = time.time()
for j in range(len(r)):
    sol = solve_ivp(sistema_1, tspan, x0,args=(r[j],b))
    t = sol.t
    x = sol.y
    for i in range(x0.size):
        plt.scatter(r[j],x[i,-1] , color='green', marker='.')

x0 = np.linspace(-.1,.1,5)
tspan = [0, -100]

r = np.linspace(0.1,2,50)
tic = time.time()
for j in range(r.size):
    sol = solve_ivp(sistema_1, tspan, x0,args=(r[j],b))
    t = sol.t
    x = sol.y
    for i in range(x0.size):
        plt.scatter(r[j],x[i,-1] , color='red', marker='.')
        plt.title('Diagrama de bifurcación')
        plt.xlabel('r')
        plt.ylabel('x')

toc = time.time()
plt.grid()
plt.show()
```
