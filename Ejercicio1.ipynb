##                                                           MÉTODO DE EULER Y RK4
### SOLUCIÓN DE ECUACIÓN DIFERENCIAL 
\begin{equation}
 \frac{dx}{dt}=x^{1/3}
\end{equation} 

#### Función para método de Euler

```
import matplotlib.style
matplotlib.style.use('classic')
import numpy as np
from matplotlib import pyplot as plt
```

```
def euler( derivadas, h, tini, tfin, x0):
    ti = tini  
    xi = x0
  
    x = list()
    t = list()
  
    while ( ti < tfin ):
    
      x.append (xi)
      
      t.append (ti)
  
      xi = xi + h*derivadas(ti,xi)
      ti = ti + h
    return t, x 
```

#### Función para método RK

```
def rk4( derivadas, h, tini, tfin, x0):
    xi = x0
    ti = tini 
    
    x = list()
    t = list()
  
    while ( ti < tfin ):
        
        t.append (ti)
        x.append (xi)
        k1 = derivadas( ti,xi)
        k2 = derivadas( ti + 0.5*h,xi + 0.5*k1*h)
        k3 = derivadas( ti + 0.5*h,xi + 0.5*k2*h)
        k4 = derivadas( ti + h,xi + k3*h) 

        xi = xi + h*(k1 + 2*k2 + 2*k3 + k4)/6
        ti = ti + h
    return t, x
```

#### Inializacón del sistema

```
def sistema_1(t, x):
    x_dot = pow(x,(1/3))
    return x_dot
```

```
h = 0.1
tini = -10
tfin = 10
x0 = 1

t1,x1 = euler( sistema_1, h, tini, tfin, x0)
t2,x2 = rk4( sistema_1, h, tini, tfin, x0,)

t = np.linspace(-12,10,1001) 
x = (pow(2/3*(t+10),(3/2)))

```

#### La solución Análitica de la ecuaion diferencial es:
\begin{equation}
x= {(\frac{2}{3}t+c)}^{\frac{3}{2}}
\end{equation} 

```
plt.figure()
# gráfica de analítica:
plt.plot(  t, x, 'black', label = 'Analítica', linewidth = 2)
# gráfica de euler:
plt.scatter( t1, x1, c = 'crimson', label = 'Euler', s = 50)
# gráfica de rk4:
plt.scatter( t2, x2, c = 'royalblue', label = 'RK4', s = 10)
plt.title('Gráfica de soluciones ecuación diferenial')
plt.xlim([-10, tfin])
plt.xlabel(r'$Tiempo$')
plt.grid()
plt.legend(loc='best')
plt.show()
```

## Crecimiento poblacional
#####SOLUCIÓN DE ECUACIÓN DIFERENCIAL 
\begin{equation}
 \frac{dN}{dt}=rN(1-\frac{N}{K})
 \end{equation} 

###FUNCION 

```
def euler1( derivadas, h, tini, tfin, x0, r,k):
    ti = tini  
    xi = x0
  
    x = list()
    t = list()
  
    while ( ti < tfin ):
    
      x.append (xi)
      
      t.append (ti)
  
      xi = xi + h*derivadas(ti,xi,r,k)
      ti = ti + h
    return t, x 
def rk41( derivadas, h, tini, tfin, x0, r, k):
    xi = x0
    ti = tini 
    
    x = list()
    t = list()
  
    while ( ti < tfin ):
        
        t.append (ti)
        x.append (xi)
        k1 = derivadas( ti,xi,r,k)
        k2 = derivadas( ti + 0.5*h,xi + 0.5*k1*h,r,k)
        k3 = derivadas( ti + 0.5*h,xi + 0.5*k2*h,r,k)
        k4 = derivadas( ti + h,xi + k3*h,r,k) 

        xi = xi + h*(k1 + 2*k2 + 2*k3 + k4)/6
        ti = ti + h
    return t, x
```

```

def sistema_1(t, x, r, k):
    x_dot = r*x*(1-(x/k))
    return x_dot

h = 0.1
tini = 0
tfin = 15
x0 = 2
r=0.5
k=100

t1,N1 = euler1( sistema_1, h, tini, tfin, x0, r, k)
t2,N2 = rk41( sistema_1, h, tini, tfin, x0, r, k)



```

###Graficas

```
plt.figure()
# gráfica de analítica:
#plt.plot(  t, x, 'black', label = 'Analítica', linewidth = 2)
# gráfica de euler:
plt.scatter( t1, N1, c = 'crimson', label = 'Euler', s = 50)
# gráfica de rk4:
plt.scatter( t2, N2, c = 'royalblue', label = 'RK4', s = 10)
plt.title('Gráfica de soluciones ecuación diferenial')
plt.xlim([0, tfin])
plt.xlabel(r'$Tiempo$')
plt.grid()
plt.legend(loc='best')
plt.show() 

for i in range(0,200,10):
    x0=i
    #r=i
    t3,x3 = euler1( sistema_1, h, tini, tfin, x0, r, k)
    t4,N4 = rk41( sistema_1, h, tini, tfin, x0, r, k)
    plt.plot( t3, x3, c = 'crimson')
    plt.plot( t4, N4, c = 'royalblue')
plt.xlim([0, 15])
#plt.ylim([-3.14, 6.28])
plt.xlabel(r'$tiempo$')
plt.title('Gráfica de soluciones ecuación diferenial con diferentes x0')
plt.grid()
plt.show()
```
